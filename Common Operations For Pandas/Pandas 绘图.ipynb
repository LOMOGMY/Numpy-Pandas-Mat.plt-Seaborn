# Pandas 绘图

```python
%matplotlib inline
import pandas as pd
import numpy as np
```

### Series图

```python
s = pd.Series(np.random.randn(10),index = np.arange(0,100,10))
# np.random.randn(10)生成了10个数据
s.plot()
```

### DataFrame图

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0), 
               index = np.arange(0, 100, 10), 
               columns = ['A', 'B', 'C', 'D'])
df.head()
```

```python
#折线图
df.plot()
```

### Series柱形图

```python
#柱形图
import matplotlib.pyplot as plt
fig,axes = plt.subplots(2,1,figsize = (10,8))
data = pd.Series(np.random.rand(16),index=list('abcdefghijklmnop'))
data.plot(ax = axes[0],kind='bar')
data.plot(ax = axes[1],kind='barh')
```

### DataFrame柱形图

```python
df = pd.DataFrame(np.random.rand(6, 4), 
               index = ['one', 'two', 'three', 'four', 'five', 'six'], 
               columns = pd.Index(['A', 'B', 'C', 'D'], name = 'Genus'))
df  
```

```python
df.plot(kind='bar')
```

### DataFrame直方图

### DataFrame散点图

```python
macro = pd.read_csv('macrodata.csv')
macro.head()
```

#### 画出所有变量的散点图矩阵，会出现变量太多，图很乱的问题，示例↓
* pd.plotting.scatter_matrix( DataFrame,color = ,alpha = )

```python
pd.plotting.scatter_matrix(macro,color='g',alpha=0.3)
```
