### 多个线条

```python
numpy = np.arange(0,10,0.5)
plt.plot(numpy,numpy,'r--')
plt.plot(numpy,numpy**2,'bs')
plt.plot(numpy,numpy**3,'go')
```

```python
# 第二种方法，一张图中画多条线
plt.plot(numpy,numpy,'r--',
        numpy,numpy**2,'bs',
        numpy,numpy**3,'go')  
```

```python
x = np.linspace(-10,10)
y = np.sin(x)
plt.plot(x,y,linewidth = 2.0)  #linewidth指定线条宽度
```

#### plt.plot(x,y,color = ,linestyle = ,marker = ,markerfacecolor = ,markersize = )

```python
plt.figure(figsize=(8,6))
plt.plot(x,y,color = 'b',linestyle = ':',marker = 'o',markerfacecolor = 'r',markersize = 10)
#marker就是标志点即np.linspace(-10,10)这些值,markersize即指定这些点的大小
```

#### 也可以先把图画出来再用 plt.setp()设置参数

```python
line = plt.plot(x,y)
plt.setp(line,color='r',linewidth = 2.0, alpha = 0.1) #alpha即透明程度
```

## 子图

## 给图加注释
* plt.annotate()

```python
plt.figure(figsize=(8,6))
plt.plot(x,y,color='b',linestyle=':',marker = 'o',markerfacecolor='r',markersize = 10)
plt.xlabel('x:---')
plt.ylabel('y:---')
plt.title('Title here:---')
plt.text(0,0,'example')
plt.grid(True)
plt.annotate('LOMO',xy=(-5,0),xytext=(-2,0.3),arrowprops = dict(facecolor='red',
                                                                shrink=0.05,headlength= 20,headwidth = 20))
#xy为箭头指向的位置 ，xytext为标注文本的位置（箭头的尾部也在这里），dict中shrink是箭头缩短的长度(0为不缩短，
#1为缩到最短)
```

# 更多细节设置

> 当前的图表和子图可以使用plt.gcf()和plt.gca()获得，分别表示Get Current Figure和Get Current Axes。在pyplot模块中，许多函数都是对当前的Figure或Axes对象进行处理，比如说：plt.plot()实际上会通过plt.gca()获得当前的Axes对象ax，然后再调用ax.plot()方法实现真正的绘图。

```python
x = range(10)
y = range(10)

fig = plt.gca()
plt.plot(x,y)

fig.axes.get_xaxis().set_visible(False) #设置x轴不可见
#fig.axes.get_yaxis().set_visible(False)
```

### 绘制一个直方图

```python
import math
x = np.random.normal(loc = 0.0,scale=1.0,size=300)
width = 0.5
bins = np.arange(math.floor(x.min())-width,math.ceil(x.max())+width,width)  
#bins就是所需要设置的间隔，画的图像表示在这些间隔里面所包含的值的个数
ax = plt.subplot(111) #只有一个图的情况下，就不用plt.subplots()了

ax.spines['top'].set_visible(False) #控制轴的属性，希望上面的轴是没有的
ax.spines['right'].set_visible(False) #控制轴的属性，希望右面的轴是没有的

plt.tick_params(bottom='off',top='off',left = 'on',right='off')
#设置刻度是否可见(刻度值还是有的)，对比下图中的下边和左边
plt.grid() #使得网格可见
plt.hist(x,alpha = 0.5, bins = bins)
```

### 当label比较长的时候可以旋转
* ax.set_xticklabels()

```python
x = range(10)
y = range(10)

labels = ['LOMO_GMY' for i in range(10)]
fig,ax = plt.subplots(figsize = (8,6))
plt.plot(x,y)
plt.title('LOMO_GMY')
ax.set_xticklabels(labels,rotation = 45,horizontalalignment='right')
# rotation设置label旋转角度， horizontalalignment='right'意思为label向右对齐
```

### legend()

```python
x = np.arange(10)
plt.figure(figsize = (8,6))
for i in range(1,4):
    plt.plot(x,i*x**2,label = 'Group %d'%i) #label表明这条线所属什么
plt.legend(loc='best') 
#如果只用plt.legend() 则标注会在右上角，会挡住曲线，所以可以加上参数 loc = 'best' 自动选择最好的放置位置
```

### 标注放在图外

```python
fig = plt.figure()
ax = plt.subplot(111)

x = np.arange(10)
for i in range(1,4):
    plt.plot(x,i*x**2,label = 'Group %d'%i)
ax.legend(loc='upper center',bbox_to_anchor = (0.5,1.15) ,ncol=3)
#bbox_to_anchor为指定标注的位置，ncol理解为标注有3列
```

```python
fig = plt.figure()
ax = plt.subplot(111)

x = np.arange(10)
for i in range(1,4):
    plt.plot(x,i*x**2,label = 'Group %d'%i)
ax.legend(loc='upper center',bbox_to_anchor = (1.15,1) ,ncol=1)
```

### 指标变透明
