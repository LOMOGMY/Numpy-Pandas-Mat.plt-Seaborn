# 3D图绘制
* from mpl_toolkits.mplot3d import Axes3D

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig) #将2维的图转化为3维的图

x = np.arange(-4,4,0.25)
y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(x,y) #画三维的图需要使得坐标变成网格

Z = np.sin(np.sqrt(X**2+Y**2))
ax.plot_surface(X,Y,Z,rstride = 1,cstride = 1,cmap='rainbow')
#rstride和cstride相当于是指定步长，也就是指定格子的个数(实际上就是指定小格的密度)，cmap='rainbow'指定为彩虹色
ax.contour(X,Y,Z,zdim='z',offset = -2 ,cmap='rainbow') 
#指定投影轮廓位置，offset=-2即向下移动2个单位

ax.set_zlim(-2,2)
plt.show()
```

### 画三维图的坐标轴

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
plt.show()
```

除了上述方法(ax = fig.add_subplot(111,projection = '3d'))也可以这么指定
* ax = fig.gca(projection='3d')

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.plot(x,y,z)
plt.show()
```

### 散点图

```python
np.random.seed(1)
def randrange(n,vmin,vmax):
    return (vmax-vmin)*np.random.rand(n)+vmin

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111,projection = '3d')
n = 100
for c,m,zlow,zhigh in [('r','o',-50,-25),('b','x','-30','-5')]:
    xs = randrange(n,23,32)
    ys = randrange(n,0,100)
    zs = randrange(n,int(zlow),int(zhigh))
    ax.scatter(xs,ys,zs,c=c,marker=m)
plt.show()
```

### 调整角度
* ax.view_init(elev = ,azim = )
其中elev指定上下旋转的角度，zim指定左右旋转的角度

```python
np.random.seed(1)
def randrange(n,vmin,vmax):
    return (vmax-vmin)*np.random.rand(n)+vmin


fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111,projection = '3d')
n = 100
for c,m,zlow,zhigh in [('r','o',-50,-25),('b','x','-30','-5')]:
    xs = randrange(n,23,32)
    ys = randrange(n,0,100)
    zs = randrange(n,int(zlow),int(zhigh))
    ax.scatter(xs,ys,zs,c=c,marker=m)
    
ax.view_init(elev=0, azim=60)
plt.show()
```

### 3D条形图

```python
fig = plt.figure(figsize=(10,8))  
ax = fig.add_subplot(111, projection='3d') 

for c, z in zip(['r', 'g', 'b', 'y'], [30, 20, 10, 0]): 
    xs = np.arange(20)
    ys = np.random.rand(20)
    cs = [c]*len(xs)
    ax.bar(xs,ys,zs = z,zdir='y',color = cs,alpha = 0.5)
    # 注意这个例子中的zdir='y'即以y轴传入的数据(也就是ys)做z轴
plt.show()
```
