# Matplotlib+Pandas+Sklearn

### 构造数据

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
df = pd.DataFrame({'Condition 1': np.random.rand(20),
                   'Condition 2': np.random.rand(20)*0.9,
                   'Condition 3': np.random.rand(20)*1.1})
df.head()
```

#### 对数据做堆叠

```python
fig,ax = plt.subplots()
df.plot.bar(ax=ax,stacked=True) #使用堆叠
plt.show()
```

#### 不对数据做堆叠

```python
fig,ax = plt.subplots()
df.plot.bar(ax=ax,stacked=False) #不使用堆叠
plt.show()
```

### 画各个指标所占的百分比

```python
from matplotlib.ticker import FuncFormatter

df_ratio = df.div(df.sum(axis=1),axis=0) #先求一个百分比

fig,ax = plt.subplots()
df_ratio.plot.bar(ax=ax,stacked=True) #画百分比条形图
ax.yaxis.set_major_formatter(FuncFormatter(lambda y,_:'{:.0%}'.format(y)))
plt.show()
```

# 一个实例

### 从网上获取数据

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00383/risk_factors_cervical_cancer.csv'
df = pd.read_csv(url, na_values="?") # na_values="?"的意思是这个数据里面所有的'?'都认为是缺失值
df.head()
```

### 对缺失值进行填充

```python
from sklearn.preprocessing import Imputer  #Imputer可以直接按照均值填充缺失值

impute =  pd.DataFrame(Imputer().fit_transform(df))  
impute.columns = df.columns
impute.index = df.index

impute.head()
```

```python
%matplotlib notebook
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
```

```python
features = impute.drop('Dx:Cancer', axis=1) #"Dx:Cancer"这列表示这个人是否患癌症，作为y值，将他去掉
y = impute["Dx:Cancer"]  #"Dx:Cancer"这列表示这个人是否患癌症，作为y值

pca = PCA(n_components=3)  #将36维的数据降维到3维
X_r = pca.fit_transform(features)  #将36维的数据降维到3维

print("Explained variance:\nPC1 {:.2%}\nPC2 {:.2%}\nPC3 {:.2%}"
      .format(pca.explained_variance_ratio_[0],
              pca.explained_variance_ratio_[1],
              pca.explained_variance_ratio_[2]))

fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(X_r[:, 0], X_r[:, 1], X_r[:, 2], c=y, cmap=plt.cm.coolwarm)  #展示这选出来的3个维度的值

# Label the axes
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

plt.show()
```
