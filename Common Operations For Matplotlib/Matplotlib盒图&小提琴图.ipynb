# 盒图&小提琴图

![](./img/intro_boxplot.png)

>离群点定义为小于Q1-1.5*IQR或大于Q3+1.5*IQR,其中IQR为Q3 - Q1

```python
%matplotlib inline
```

* plt.boxplot(data,notch,sym,vert)

data:数据

notch：指定盒图的样子，bool变量(有两种情况可选)

sym：指定离群点用什么来表示，'s'表示用方框表示

vert:垂直画

```python
import matplotlib.pyplot as plt
import numpy as np

data = [np.random.normal(0,std,100) for std in range(1,4)]
fig = plt.figure(figsize = (8,6))
plt.boxplot(data,notch=False,sym='s',vert=True) 

plt.xticks([y+1 for y in range(len(data))],['x1','x2','x3']) #相当于先指定刻度[1,2,3]的位置，再改为x1,x2,x3
plt.xlabel('x')
plt.title('box plot')
```

### 设置颜色

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
fig = plt.figure(figsize = (8,6))
bplot = plt.boxplot(data,notch=False,sym='s',vert=True)

plt.xticks([y+1 for y in range(len(data))],['x1','x2','x3'])
plt.xlabel('x')
plt.title('box plot')

for components in bplot.keys():
    for line in bplot[components]:
        line.set_color('blue')
```

### 横着画

只需要将上面的xticks和xlabel改为yticks和ylabel

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
fig = plt.figure(figsize = (8,6))
plt.boxplot(data,notch=False,sym='s',vert=False)

plt.yticks([y+1 for y in range(len(data))],['x1','x2','x3'])
plt.ylabel('x')  #这里可改为plt.ylabel('y') 
plt.title('box plot')
```

### nortch=True

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
fig = plt.figure(figsize = (8,6))
plt.boxplot(data,notch=True,sym='s',vert=False)

plt.yticks([y+1 for y in range(len(data))],['x1','x2','x3'])
plt.ylabel('x')  #这里可改为plt.ylabel('y') 
plt.title('box plot')
```

### 向图里面填充颜色
*  plt.boxplot(.... patch_artist=True)
* .set_facecolor()


```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
fig = plt.figure(figsize = (8,6))
bplot = plt.boxplot(data,notch=False,sym='s',vert=True,patch_artist=True) 
#必须指定patch_artist=True才能进行颜色填充

plt.xticks([y+1 for y in range(len(data))],['x1','x2','x3'])
plt.xlabel('x')
plt.title('box plot')

colors = ['pink','lightblue','lightgreen']
for pathch,color in zip(bplot['boxes'],colors):
    pathch.set_facecolor(color)
```

# 小提琴图（Violinplot）
* .biolinplot()

小提琴图和盒图的对比

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(12,5))
data = [np.random.normal(0,std,100) for std in range(6,10)]
axes[0].violinplot(data,showmeans=False,showmedians=True) #是否展示平均数或中位数
axes[0].set_title('violin plot')

axes[1].boxplot(data)
axes[1].set_title('box plot')

for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y+1 for y in range(len(data))]) #先指定[1,2,3,4]的位置
plt.setp(axes,xticks=[y+1 for y in range(len(data))],xticklabels=['x1','x2','x3','x4'])  
#再将[1,2,3,4]指定为[x1,x2,x3,x4]
```
