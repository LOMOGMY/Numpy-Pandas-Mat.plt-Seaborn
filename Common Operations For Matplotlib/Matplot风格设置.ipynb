# Matplot风格设置

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### 展示所有可用风格

```python
plt.style.available  
```

```python
x = np.linspace(-10,10)
y = np.sin(x)
plt.plot(x,y)
```

### 设置风格
* plt.style.use()

```python
plt.style.use('dark_background')
plt.plot(x,y)
```

```python
# 可以混合风格，但是不推荐
plt.style.use(['ggplot','bmh'])
plt.plot(x,y)
```
