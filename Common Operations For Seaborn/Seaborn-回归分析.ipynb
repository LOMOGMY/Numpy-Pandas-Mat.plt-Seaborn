# Seaborn回归分析

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
sns.set(color_codes=True)

np.random.seed(sum(map(ord, "regression")))

tips = sns.load_dataset("tips")

tips.head()
```

### regplot()和lmplot()都可以绘制回归关系,推荐regplot()

lmplot()里面有更高级的内容

```python
plt.figure(figsize=(8,6))
sns.regplot(x="total_bill", y="tip", data=tips)  #这里"total_bill"和"tip"是data的column值
```

```python
plt.figure(figsize=(8,6))
sns.lmplot(x="total_bill", y="tip", data=tips) 
```

### 查看size和tip之间的关系

```python
plt.figure(figsize=(8,6))
sns.regplot(data=tips,x="size",y="tip")
```

### 上面这种情况不适合做回归分析，若要做，则需要对数据进行小的抖动
* sns.regplot( x_jitter= )

```python
sns.regplot(x="size", y="tip", data=tips, x_jitter=.05) #原始x只能是整数，这里加上了小范围的抖动
```
