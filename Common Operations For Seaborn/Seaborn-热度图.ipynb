# Seaborn热度图

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np; 
np.random.seed(0)
import seaborn as sns;
sns.set()
```

### 热度矩阵示例

```python
uniform_data = np.random.rand(3, 3)
print (uniform_data)
heatmap = sns.heatmap(uniform_data)
```

### 控制右边调色板(color bar)的取值范围
* sns.heatmap( , vmin=, vmax=)

```python
ax = sns.heatmap(uniform_data, vmin=0.2, vmax=0.5)
```

### 定义center=，可以指定中心值

```python
normal_data = np.random.randn(3, 3)
print (normal_data)
ax = sns.heatmap(normal_data, center=0)
```

## 一个实例

```python
flights = sns.load_dataset("flights")
flights.head()
```

### 可让结果一目了然

```python
flights = flights.pivot("month", "year", "passengers") #定义横轴为年份，纵轴为月份
print (flights)
plt.figure(figsize=(8,6))
ax = sns.heatmap(flights)
```

```python
plt.figure(figsize=(8,6))
ax = sns.heatmap(flights, annot=True,fmt="d") 
#fmt='d'是指定字体的一种格式，若用默认格式则会用科学计数法表示，且会引起数字之间的重叠
```

### 也可以自定义格之间的间距

```python
plt.figure(figsize=(8,6))
ax = sns.heatmap(flights, linewidths=.5)
```

### 指定自己的color bar

```python
plt.figure(figsize=(8,6))
ax = sns.heatmap(flights, cmap="YlGnBu")  
```

```python
ax = sns.heatmap(flights, cbar=False) #可以隐藏color bar，但是不推荐
```
