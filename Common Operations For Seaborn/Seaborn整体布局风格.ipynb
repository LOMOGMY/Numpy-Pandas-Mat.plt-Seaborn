# Seaborn-Style

```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

<img src="f1.png" alt="FAO" width="990">

```python
# matplotlib
figure = plt.figure(figsize=(8,6))
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sinplot()
```

## Seaborn5

```python
figure = plt.figure(figsize=(8,6))
sns.set() #Seaborn
sinplot()
```
