# Seaborn-Style

```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

<img src="f1.png" alt="FAO" width="990">

```python
# 用matplotlib画图示例
figure = plt.figure(figsize=(8,6))
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sinplot()
```

## Seaborn这个库自带有5个模板

```python
figure = plt.figure(figsize=(8,6))
sns.set() #无参数意味着用Seaborn默认的参数组合
sinplot()
```

### 5种主题风格 ###
* darkgrid
* whitegrid
* dark
* white
* ticks

```python
sns.set_style("whitegrid")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

### 若不想要刻度线，可用sns.set_style("dark")或sns.set_style("white")

```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

### 标明刻度

```python
sns.set_style("ticks")
sinplot()
```

### 只保留x轴和y轴(去除上边和右边的线)
* sns.despine()

```python
sinplot()
sns.despine()
```

#### sns.despine(offset=10), offset可以指定图形离轴线的位置

```python
#f, ax = plt.subplots()
sns.violinplot(data)
sns.despine(offset=10)
```

#### 可以利用sns.despine(left=True)隐藏轴(但是轴上的值还是有的)

```python
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
sns.despine(left=True)
```

### 绘制子图&不同的风格
* with sns.axes_style("darkgrid"):

```python
plt.figure(figsize=(8,6))
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

### 设定绘图域的大小和线条的大小

线条大小：
* sns.set_context()  

```python
sns.set()
sns.set_context("paper")   #线条较细
plt.figure(figsize=(6,4))
sinplot()
```

```python
sns.set_context("talk")  #线条较paper粗
plt.figure(figsize=(6,4))
sinplot()
```

```python
sns.set_context("poster")  #线条较talk粗
plt.figure(figsize=(6,4))
sinplot()
```

### font_scale指定字体大小，lines.linewidth指定线条粗度

```python
sns.set_context("notebook",font_scale=2.0, rc={"lines.linewidth": 4.5})
sinplot()
```
